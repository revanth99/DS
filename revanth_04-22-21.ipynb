```python
import numpy as np
import pandas as pd
```

```python
filename="connecticut-history.csv"
fileURl = "https://covidtracking.com/data/download/"

dfCT = pd.read_csv(fileURl+filename)


filename1="colorado-history.csv"
fileURl = "https://covidtracking.com/data/download/"

dfCO = pd.read_csv(fileURl+filename1)
```

```python
# 1  the number of rows of data, and the number of columns for connecticut
rows, columns=dfCT.shape
print("number of rows of data:",rows)
print("number of columns:", columns)
print(dfCT.head(3))
print(dfCT.tail(3))
print(dfCT.head(3).T)
print(dfCT.tail(3).T)
```

```python
# 1  the number of rows of data, and the number of columns for colorado
rows, columns=dfCO.shape
print("number of rows of data:",rows)
print("number of columns:", columns)
print(dfCO.head(3))
print(dfCO.tail(3))
print(dfCO.head(3).T)
print(dfCO.tail(3).T)
```

```python
#2a Are there any columns that have ALL NaNs?
print("Are there any columns that have ALL NaNs in connecticut-history.csv :",dfCT.isnull().values.any())
print("Are there any columns that have ALL NaNs in colorado-history.csv :",dfCO.isnull().values.any())
```

```python
#2b Columns having ALL NaNs in connecticut
print("Columns which have all NaNs for connecticut\n", dfCO.isnull().all(axis=0)[lambda x:x])
```

```python
print("Columns which have all NaNs for colorado\n", dfCO.isnull().all(axis=0)[lambda x:x])
```

```python
#3a Number of non NaN values of each row in connecticut-history.csv
dfCT = dfCT.T
A = dfCT.dropna(axis='columns', how='all')
c =A.count()
print("the number of non-Nan values for each row in connecticut\n",A.count())
print("the maximum of these numbers is ", A.count().max())

#3b  rows having maximum number of values in connecticut
print("rows having maximum number of values in connecticut",(c == A.count().max()).sum())
```

```python
#3a Number of non NaN values of each row in colorado-history.csv
dfCO = dfCO.T
A = dfCO.dropna(axis='columns', how='all')
d =A.count()
print("the number of non-Nan values for each row in colorado\n",A.count())
print("the maximum of these numbers is ", A.count().max())

#3b rows having maximum number of values in colorado
print("rows having maximum number of values in colorado",(d == A.count().max()).sum())
```

```python
#4a Number of non NaN values of each row in connecticut-history.csv
A = dfCT.dropna(axis='columns', how='all')
c =A.count()
print("the number of non-Nan values for each row in connecticut\n",A.count())
print("the minimum of these numbers is ", A.count().min())

#4b  rows having minimum number of values in connecticut
print("rows having minimum number of values in connecticut",(c == c.min()).sum())
```

```python
#4a Number of non NaN values of each row in colorado-history.csv
dfCO_notNaN = dfCO.dropna(axis='columns', how='all')
d =dfCO_notNaN.count()
print("the number of non-Nan values for each row\n",d)
print("the minimum of these numbers is ", d.min())

#4b rows having minimum number of values
print("rows having minumum number of values",(d == d.min()).sum())
```

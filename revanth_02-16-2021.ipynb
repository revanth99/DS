```python
# Exercise 10.1
def nested_sum(t):
    total = 0
    for n in t:
        total += sum(n)
    return total

t = [[1, 2], [3], [4, 5, 55]]
nested_sum(t)
```

```python
# Exercise 10.2
def cumsum(t):
    total = 0
    cum= []
    for x in t:
        total+= x
        cum.append(total)
    return cum
        

t = [1, 2, 3, 4, 5, 55]
cumsum(t)
```

```python
# Exercise 10.3
def middle(t):
    return t[1:-1]
    
    
t = [1, 2, 3, 4, 5, 55]
middle(t)
```

```python
# Exercise 10.5
def is_sorted(t):
    return t == sorted(t)

t = [1, 2, 3, 4, 5, 55]
is_sorted(t)
```

```python
# Exercise 10.6
def is_anagram(w1,w2):
    return sorted(w1) == sorted(w2)

t1=['r','e','v']
t2=['e','v','r']
is_anagram(t1,t2)
```

```python
# Exercise 10.9
def word_list1():
    t = []
    a = open('words-1.txt')
    for l in a:
        word = l.strip()
        t.append(word)
    return t
```

```python
def word_list2():
    t = []
    a = open('words-1.txt')
    for l in a:
        word = l.strip()
        t = t + [word]
    return t
```

```python
# exercise 10.11
def reverse_pair():
    txt=open('words-1.txt')
    for l in txt:
        l=l.replace('\n','')
        rev_line= l[::-1]
        if(l==rev_line):
            print(l)
```

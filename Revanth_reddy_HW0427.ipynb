```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename="connecticut-history.csv"
fileURl = "https://covidtracking.com/data/download/"

dfCT = pd.read_csv(fileURl+filename)


filename1="colorado-history.csv"
fileURl = "https://covidtracking.com/data/download/"

dfCO = pd.read_csv(fileURl+filename1)
```

```python

plt.plot(dfCT.positive,dfCT.totalTestResults)
plt.xlabel('No of positive cases', fontsize=18)
plt.ylabel('Total No of tests', fontsize=16)
plt.title('positive cases vs Total No of tests in Connecticut',fontsize=18)
plt.show()
```

```python
plt.plot(dfCT.positive,dfCT.death)
plt.xlabel('No of positive cases', fontsize=18)
plt.ylabel('No of deaths', fontsize=16)
plt.title('positive cases vs Deaths in Connecticut',fontsize=18)
plt.show()
```

```python
plt.plot(dfCO.positive,dfCO.death)
plt.xlabel('No of positive cases', fontsize=18)
plt.ylabel('No of deaths', fontsize=16)
plt.title('positive cases vs Deaths in Colorado',fontsize=18)
```
